```python
def minmaxnorm(T):
    maxi = torch.max(T,0)[0]
    mini = torch.min(T,0)[0]
    T = (T-mini)/(maxi-mini)
    return(T)


def replace_heure_d_ete(tensor,start = 572, end = 576):
    values_before = tensor[start-1:start]
    values_after = tensor[end:end+1]

    mean_values = (values_before + values_after) / 2
    mean_values = mean_values.repeat(4,1,1,1)

    tensor[start:end] = mean_values
    return tensor

def find_positions(applications, file_list):
    positions = []
    for app in applications:
        for idx, file_path in enumerate(file_list):
            if app in file_path:
                positions.append(idx)
    return positions


def plot_profil_app(T,apps,selected_apps=['Facebook_Messenger','Snapchat','WhatsApp'],transfer=None,name = 'AMP',norm=False,x_lim = 672,y_lim=None):
    df=pd.DataFrame()
    for app in selected_apps: #music_apps #trafic_apps # direct_messenger_apps # social_networks_apps #['Uber','Spotify','WhatsApp','Instagram']
        selected_apps = [app]
        positions = find_positions(selected_apps,apps)
        trafic_pos = [2*k for k in positions] + [2*k+1 for k in positions]
        if norm:
            T_dl = minmaxnorm(torch.mean(T[:,trafic_pos[0],:,:],[1,2]))
            T_up = minmaxnorm(torch.mean(T[:,trafic_pos[1],:,:],[1,2]))
        else:
            T_dl = torch.mean(T[:,trafic_pos[0],:,:],[1,2])
            T_up = torch.mean(T[:,trafic_pos[1],:,:],[1,2]) 
        if app == 'Facebook_Messenger': 
            app = 'Messenger'
        if app == 'Apple_iMessage':
            app = 'iMessage'
        if transfer is None:
            df[f'dl_{app}'] = T_dl
            df[f'up_{app}'] = T_up
        elif transfer == 'DL':
            df[f'dl_{app}'] = T_dl
        elif transfer == 'UP':
            df[f'up_{app}'] = T_up
        else:
            raise NotImplementedError(f'transfer {transfer} has not been implemented')

    #pd.DataFrame({f'dl_{app}':T_dl,f'up_{app}':T_up}).plot()
    df.plot(alpha = 0.8)
    plt.xlim([0,x_lim])
    if y_lim is None:
        if norm:
            y_lim = 1
        else:
            y_lim = df.max().max()/2
            
    plt.ylim([0,y_lim])        
    plt.title(f"Profil des intensité de consommation d'application mobile à 2km autour de la station {name}")
    display(df.corr())
    
    return df
    
```

```python
from ray import tune

# Espace de recherche conditionnel structuré
def search_space():
    return {
        "a": tune.choice([
            {   # Si a == 1
                "a": 1,
                "b": tune.choice([4, 5])  # b est choisi parmi [0, 1]
            },
            {   # Si a == 2
                "a": 2,
                "b": tune.choice([3, 4])  # b est choisi parmi [2, 3]
            },
            {   # Si a == 3
                "a": 3,
                "b": tune.choice([1, 2])  # b est choisi parmi [2, 3]
            }
        ])
    }

# Fonction d'optimisation (objectif à minimiser)
def objective(config):
    a = config["a"]["a"]
    b = config["a"]["b"]
    
    result = a + b  # On veut minimiser a + b
    tune.report(result=result)

# Exécution de l'optimisation avec Ray Tune
analysis = tune.run(
    objective,  # Fonction objectif
    config=search_space(),  # Espace de recherche conditionnel
    num_samples=10,  # Nombre d'échantillons à tester
    metric="result",  # Metric à optimiser (minimiser)
    mode="min",  # Minimiser la metric
)

# Affichage de la meilleure configuration trouvée
print("Meilleure configuration trouvée:", analysis.best_config)
print("Valeur minimale de a + b:", analysis.best_result['result'])

```
