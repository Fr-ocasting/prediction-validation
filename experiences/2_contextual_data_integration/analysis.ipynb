```python
import os 
import pandas as pd 
from pipeline.plotting.TS_analysis import plot_TS


def load_csv(exp_i,trial_j,n_bis,df_j_all,metric_i,columns_metrics):
    local_folder_path =  f"{folder_path}/{exp_i}/{dic_exp_to_names[exp_i]}/{trial_j}_bis{n_bis}"
    Losses_file_path = f"{local_folder_path}/Losses_{trial_j}_bis{n_bis}.csv"
    df_j = pd.read_csv(Losses_file_path,index_col = 0)
    df_j.columns = [f'Train_{n_bis}',f'Valid_{n_bis}']
    df_j_all = pd.concat([df_j_all,df_j],axis=1)

    # Get metrics : 
    metrics_ij = pd.read_csv(f"{local_folder_path}/METRICS_{trial_j}_bis{n_bis}.csv",index_col = 0).loc['test',columns_metrics]
    metric_i.append(metrics_ij)
    return df_j_all, metric_i


def tackle_trial_j(L_metrics,exp_i,trial_j,columns_metrics,plot_losses):
    df_j_all = pd.DataFrame()
    metric_i = []
    for n_bis in range(1,6):

        df_j_all, metric_i = load_csv(exp_i,trial_j,n_bis,df_j_all,metric_i,columns_metrics)

    metric_i = pd.DataFrame(pd.DataFrame(metric_i).agg(['mean','std']).unstack()).T
    metric_i.index = [f"{trial_j}"]
    L_metrics.append(metric_i)
    if plot_losses:
            plot_TS(df_j_all,width=1500,height=400,bool_show=True,title=f"{exp_i}, {trial_j}",x_datetime = False)
    return L_metrics

folder_path = 'save/K_fold_validation/training_wo_HP_tuning'
metrics = ['rmse','mae','mase']
plot_losses = False 



for exp_i,trials in dic_trials.items():
    print(f"\n------------------ {exp_i} ---")
    columns_metrics = [f"{m}_h{dic_exp_to_h[exp_i]}" for m in metrics]
    L_metrics = []
    for trial_j in trials:
        L_metrics = tackle_trial_j(L_metrics,exp_i,trial_j,columns_metrics,plot_losses)
    
    if exp_i == 'Exp4':
        exp_i_tmp = 'Exp3'
        trial_j = 'STAEformer_bike_out_calendar__e200_h1'
        L_metrics = tackle_trial_j(L_metrics,exp_i_tmp,trial_j,columns_metrics,plot_losses)
    if exp_i == 'Exp4_15min':
        exp_i_tmp = 'Exp3_bike_15min_h4'
        trial_j = 'STAEformer_bike_out_calendar__e200_h4'
        L_metrics = tackle_trial_j(L_metrics,exp_i_tmp,trial_j,columns_metrics,plot_losses)
    df_metrics_all = pd.concat(L_metrics)
    
    display(df_metrics_all.sort_values(by=[(f'rmse_h{dic_exp_to_h[exp_i]}','mean')]))
```

```python
def tackle_trial_j(L_metrics,exp_i,trial_j,columns_metrics):
    df_j_all = pd.DataFrame()
    metric_i = []
    for n_bis in range(1,6):

        df_j_all, metric_i = load_csv(exp_i,trial_j,n_bis,df_j_all,metric_i,columns_metrics)

    metric_i = pd.DataFrame(metric_i)
    metric_i.index = [f"{trial_j}_bis{n_bis}" for n_bis in range(1,6)]
    L_metrics.append(metric_i)

    return L_metrics

def build_legend_group_exp4(x):
    if ('adpQ0' in x) and ('adp0' in x):
        return 'adp0 & adpQ0'
    if ('adpQ0' in x):
        return 'adpQ0'
    if ('adp0' in x):
        return 'adp0'
    if not('adp' in x):
        return 'Baseline'
    return 'STAEformer_CrossAttn'

def update_df_metrics_exp4_15min(df_metrics_all):
    df_metrics_all['legend_group'] = df_metrics_all.reset_index()['index'].apply(build_legend_group_exp4).values
    df_metrics_all['id'] = [c.split('_CrossAttnBackBone_')[1].split('__')[0] if '_CrossAttnBackBone_' in c else 'Baseline' for c in df_metrics_all.index]
    df_metrics_all['id'] = df_metrics_all['id'].fillna('Baseline')
    df_metrics_all = df_metrics_all.rename(columns= {'rmse_h4':'rmse','rmse_h1':'rmse','mae_h4':'mae','mae_h1':'mae','mase_h4':'mase','mase_h1':'mase'})
    return df_metrics_all

def build_legend_group_exp1(x):
    if not('subway_in' in x) :
        return 'Baseline'
    elif 'independant_embedding' in x:
        return 'Independant Embedding'
    elif 'shared_embedding' in x:
        return 'Shared Embedding'
    else:
        return 'Other Methods'

def update_df_metrics_exp1(df_metrics_all):
    df_metrics_all['legend_group'] = df_metrics_all.reset_index()['index'].apply(build_legend_group_exp1).values
    df_metrics_all['id'] = [c.split('_calendar_')[1].split('__')[0] if 'subway_in' in c else 'Baseline' for c in df_metrics_all.index]
    df_metrics_all = df_metrics_all.rename(columns= {'rmse_h4':'rmse','rmse_h1':'rmse','mae_h4':'mae','mae_h1':'mae','mase_h4':'mase','mase_h1':'mase'})
    return df_metrics_all

def update_df_metrics(df_metrics_all,exp_i):
    if exp_i == 'Exp1':
        return update_df_metrics_exp1(df_metrics_all)
    elif exp_i == 'Exp4_15min':
       return update_df_metrics_exp4_15min(df_metrics_all)
    else:
        raise NotImplementedError





exp_i = 'Exp1' #'Exp4_15min', 'Exp4' #'Exp3_bike_15min_h4' #'Exp3' #'Exp1'
trials = dic_trials[exp_i]
print(f"\n------------------ {exp_i} ---")
columns_metrics = [f"{m}_h{dic_exp_to_h[exp_i]}" for m in metrics]
L_metrics = []
for trial_j in trials:
    L_metrics = tackle_trial_j(L_metrics,exp_i,trial_j,columns_metrics)

if exp_i == 'Exp4':
    exp_i_tmp = 'Exp3'
    trial_j = 'STAEformer_bike_out_calendar__e200_h1'
    L_metrics = tackle_trial_j(L_metrics,exp_i_tmp,trial_j,columns_metrics)
if exp_i == 'Exp4_15min':
    exp_i_tmp = 'Exp3_bike_15min_h4'
    trial_j = 'STAEformer_bike_out_calendar__e200_h4'
    L_metrics = tackle_trial_j(L_metrics,exp_i_tmp,trial_j,columns_metrics)
df_metrics_all = pd.concat(L_metrics)



df_metrics_all = update_df_metrics(df_metrics_all,exp_i)

plot_boxplot_on_metric(df_metrics_all, metric_i='mae', xaxis_label="Config", legend_group='legend_group', width=1200, height=400, 
                            save_path=None)
```
