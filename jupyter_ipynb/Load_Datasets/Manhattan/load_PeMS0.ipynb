```python
import sys 
import os 
import pandas as pd
import numpy as np 
import h5py

current_file_path = os.getcwd()
ROOT = os.path.abspath(os.path.join(current_file_path,'..','..'))
if ROOT not in sys.path:
    sys.path.insert(0,ROOT)

from constants.paths import FOLDER_PATH
```

```python
for PeMS_num,year in zip([3,4,7,8],[2018,2018,2017,2016]):
    data_type = 'PEMS'
    print('\n-------------------------------')
    print(f"{data_type}0{PeMS_num} Dataset")
    dirname = f"{ROOT}/{FOLDER_PATH}/{data_type}0{PeMS_num}"

    adj_mx = np.load(f"{dirname}/{data_type.lower()}0{PeMS_num}_rn_adj.npy")
    data = h5py.File(f"{dirname}/{data_type.lower()}0{PeMS_num}_his_{year}.h5", 'r')

    data['t']['axis0'].shape,data['t']['axis1'].shape,data['t']['block0_items'].shape,data['t']['block0_values'].shape

    df = pd.DataFrame(data['t']['block0_values'],index = data['t']['axis1'],columns = data['t']['axis0'] )
    #print(list(data.keys()))   # ['t']  ## ['axis0', 'axis1', 'block0_items', 'block0_values']
    df.index = pd.to_datetime(df.index, unit='ns')
    print('df.shape: ',df.shape)
    print(f"START: '{df.index.min()}'")
    print(f"END: '{df.index.max()}'")
    print('Nb time-step within Coverage Period: ',len(pd.date_range(df.index.min(),df.index.max(),freq='5min')))
    display(df.head())

```
