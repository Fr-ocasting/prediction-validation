```python
import os
import glob
import pandas as pd

#init_path = 'save/STGCN/graph_conv/sym_norm_lap/act_glu_Ks2_opt_adamw/'
init_path = 'save/CNN/h_dims16_16_out_dims16_2_opt_adamw/'
columns = ['Embedding','train_prop','valid_prop','calib_prop','best_epoch','CalendarClass','Position','Specific_lr','train_loss','valid_loss','PICP','MPIW','PICP_cqr','MPIW_cqr']
results = pd.DataFrame(columns = columns)
for specific_lr in ['True','False']:
    for calendarclass in [1,2,3]:
        for position in ['input','output']:
            FOLDER_PATH = f"{init_path}Specific_lr_{specific_lr}/CalendarClass{calendarclass}/position_{position}/"
            calib_0_csv = glob.glob(f"{FOLDER_PATH}*calib0E*.csv")[0]
            calib_05_csv = glob.glob(f"{FOLDER_PATH}*calib05*.csv")[0]
           
            df_0 = pd.read_csv(calib_0_csv,index_col = 0)
            df_05 = pd.read_csv(calib_05_csv,index_col = 0)
            
            
            for df,calib_prop in zip([df_0,df_05],[0,0.5]):
                epoch,trainloss,validloss,picp,mpiw,picp_cqr,mpiw_cqr = df.values.reshape(-1)
                epoch,trainloss,validloss,picp,mpiw,picp_cqr,mpiw_cqr = int(epoch),'{:.5f}'.format(trainloss),'{:.5f}'.format(validloss),'{:.2%}'.format(picp),'{:.3f}'.format(mpiw),'{:.2%}'.format(picp_cqr),'{:.3f}'.format(mpiw_cqr)

                results.loc[len(results)] = dict(Embedding = 'On', train_prop = 0.6, valid_prop = 0.2, calib_prop = calib_prop,
                                                 best_epoch = epoch,CalendarClass = calendarclass,Position = position,
                                                 Specific_lr = specific_lr, train_loss = trainloss, valid_loss = validloss,
                                                 PICP = picp, MPIW = mpiw, PICP_cqr = picp_cqr, MPIW_cqr = mpiw_cqr)
                                                 
        
                                                 
FOLDER_PATH = f"{init_path}no_embedding/"
calib_0_csv = glob.glob(f"{FOLDER_PATH}*calib0E*.csv")[0]
calib_05_csv = glob.glob(f"{FOLDER_PATH}*calib05*.csv")[0]

df_0 = pd.read_csv(calib_0_csv,index_col = 0)
df_05 = pd.read_csv(calib_05_csv,index_col = 0)


for df,calib_prop in zip([df_0,df_05],[0,0.5]):
    epoch,trainloss,validloss,picp,mpiw,picp_cqr,mpiw_cqr = df.values.reshape(-1)
    epoch,trainloss,validloss,picp,mpiw,picp_cqr,mpiw_cqr = int(epoch),'{:.4f}'.format(trainloss),'{:.4f}'.format(validloss),'{:.2%}'.format(picp),'{:.3f}'.format(mpiw),'{:.2%}'.format(picp_cqr),'{:.3f}'.format(mpiw_cqr)

    results.loc[len(results)] = dict(Embedding = 'Off', train_prop = 0.6, valid_prop = 0.2, calib_prop = calib_prop,
                                     best_epoch = epoch,CalendarClass = None,Position = None,
                                     Specific_lr = None, train_loss = trainloss, valid_loss = validloss,
                                     PICP = picp, MPIW = mpiw, PICP_cqr = picp_cqr, MPIW_cqr = mpiw_cqr)



results.sort_values('valid_loss').to_csv(f"{init_path}results.csv")     
```
